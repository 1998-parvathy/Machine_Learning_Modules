```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Day=[1,2,3,4,5,6,7]
Temp=[50,51,52,48,47,49,46]
```

```python
plt.plot(Day,Temp)
plt.show()
```

```python
plt.plot(Day,Temp,color='green',linestyle='dashed',marker='o',linewidth=2)
plt.show()
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title("Weather Info")
plt.plot(Day,Temp,color='green',linestyle='dashdot',marker='o',linewidth=2,alpha=0.4)
plt.show()
```

```python
# Legend, Grid
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[20,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel('Day',fontsize=10)
plt.ylabel('Temperature',fontsize=10)
plt.title("Weather Info")

plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
plt.legend(loc="best",shadow=True,fontsize="small")#loc=best means it give best position
plt.grid()
plt.show()
```

```python
# Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent-1400","Food","Phone/Internet Bill","Car ","Other Utilities"]
fig=plt.figure(figsize=(10,7))
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct="%0.3f%%",shadow=True,explode=[0.1,0,0.2,0,0],startangle=45)
plt.legend()
# plt.show()
plt.savefig("piechart_new.jpg")#plt.savefig("piechart_new.pdf")
```
